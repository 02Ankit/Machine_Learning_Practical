```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent and Dependent features
X = df.iloc[:,:-1]
y = iris.target
```

```python
X,y
```

```python
## Train test Split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.30, random_state = 42)
```

```python
X_train
```

```python
# Apply Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
## By default jo bhi parameters hai wo aajate hai kyunki koi abhi parameter nahi diye hai 
# to ye Post-Prunning ho jata hai 
# but yahan hum Pre-Prunning karna hai.
# kyunki Pre-Prunning ke Assumtions Best hota hai 
# PrePrunning me jub ye Desicion Tree construct hota hai 
# usi time me ye parameters apply hokar assumtions karega wo at the time creating of decision tree valuate karta hai 
treemodel = DecisionTreeClassifier(max_depth = 2)
```

```python
treemodel.fit(X_train,y_train)
```

```python
# ab decision tree ko construct karenge 
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
## har ek tree ka binary division hoga to isse hum CART Tree bolte hai 
## by default Gini use hai or Gini hamesha 0 to 0.5 range hota hai 

# isme X[3] ka malab column petal_length ke hisab se pura graph split hua hai 
# value = [31,37,37] basically ye categories ko batata hai 
# value = [count_numberof_setosa =31, count_numberof_versicolor = 37, count_numberof_verginica =37]]
```

```python
# upar Graph se patah chal raha hai ki ye Post Prunning hai.
#matlab ki iske pure depth tak jane ki jaruwat nahi hai 
## isliye humlog is model pe maxdepth hyparparameter ka use karenge 
# taki ye bata de ki iska max depth upar ke level pe hi hai 
```

```python
# prediction 
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```

```python
# PrePrunning 
Parameters ={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2'],
    # 'ccp_alpha':[1,2,3,4,5,6,7] abhi model pe kuch overfitting aa raha hai is liye hata ke dekhte hai 
} 
```

```python
# GridSearchCV sare parameters leke train karke jo acha parameters hoga use lena hai 
from sklearn.model_selection  import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel, param_grid = Parameters, cv=5, scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
# yaha pe mujhe pura information de raha hai ki kya kya lena chahiye 
cv.best_params_
#ab check karenge ki pridiction is model se kaisa aayega 
```

```python
y_pred = cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
