```python
# ceate data Frame 
# Independent feature
X= pd.DataFrame(df_housing.data, columns = df_housing.feature_names )
# Dependent feature
y = df_housing.target
```

```python
X.head()
```

```python
## Train test Split before preprocessing bhi kar sakte hai 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.30, random_state = 42)
```

##  DecisionTreeRegressor
1.) yahan hum log  DecisionTreeRegressor le rahe hai.

2.) ye ek Regressor problem hai kyunki output features y  me continuos value hai  



```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor() 
# yahan DecisionTreeRegressor() default parameter le rae hai kyunki hum training data se hyperparameter tuning nahi karte hai 
# sirf model ko train karte hai 
```

DecisionTree by default Overfiting karta hai generally kyunki ye depth tak split karta hai 

```python
regressor.fit(X_train,y_train)# yahan pe data ko train kiya jo ki ek model dega 
```

```python
# ab regressor me jo model hai use predict karte hai X_test data ko 
y_pred = regressor.predict(X_test)  
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score  = r2_score(y_pred,y_test)
```

humara score correct hai kyunki 1 ke pass ana chahiye 
agar humare pass koi non-linear datset hai or regression pata nahi kare hota to 
to humlog non-linear functions use karte hai Decission tree ek example ho gaya.

```python
score
```

## Hyperparameter 
DecisionTreeRegressor() isme bahoot se parameters hote hai per abhi pata nahi ki koun sa use hoga, to  in parameters ko lene ke liye hume hyperparameter ko tune karna padega to iske liye hi hyperparameter tnning karne ke liye hi Validation Data seprate karna padhta hai.


DecisionTree by default Overfiting karta hai generally kyunki ye depth tak split karta hai, or agar hum or bhi parameters ko apply kar ke dekhte hai to iska accuracy or bhi acha dete jata hai.


```python
# sare parameters use karte hai hyper parameters use karne ke liye 
parameters ={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,6,7,8,10,11,12],
    'max_features': ['auto', 'sqrt', 'log2']
}
regressor = DecisionTreeRegressor()
```

 https://scikit-learn.org/stable/modules/model_evaluation.html
 
regression ke case me koun koun sa scoring(Loss Functions) le sakte hai ye above link me hai.


```python
from sklearn.model_selection import GridSearchCV
regressorcv = GridSearchCV(regressor, param_grid = parameters, cv=5, scoring ='neg_mean_squared')
```

ab X_train, y_train ko fit karenge regressorcv me to  x_train me se hi validation seprate data banaeya.
 or 5 times experiments karega 5 acuuracy ke sath

```python
regressorcv.fit(X_train, y_train) # ispe errors aate hai sklearn me 
```

```python
# regressorcv.fit(X_train, y_train)
regressorcv.best__params_
```

```python

y_pred = regressorcv.predict(X_test)
```

```python
from sklearn.metrics import r2_score 
score = r2_score(y_pred,y_test)
```
