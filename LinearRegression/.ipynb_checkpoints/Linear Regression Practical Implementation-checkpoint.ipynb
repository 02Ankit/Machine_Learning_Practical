```python
# ceate data Frame 
dataset = pd.DataFrame(df_housing.data)
```

```python
dataset.columns = df_housing.feature_names 
```

```python
# Price of the House
output = pd.DataFrame( df_housing.target)
output
```

```python
# Price of the House
output.columns = df_housing.target_names
```

```python
# Price of the House
output.head()
```

```python
# Seprate the Independent and Dependent features
X = dataset
y = output
```

```python
## Train test Split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.30, random_state = 42)
```

```python
X_test
```

```python
# Applying Linear regression 
# data ka Gradient decent curve banane ke liye Data ko Standadies karna padhta hai.
# X_train ko Standerise karenge fit transform se or  X_test ko sirf transform karenge
#Output Feature y_train and y_test ko nahi karna chahiye
```

## https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
The standard score of a sample x is calculated as:

z = (x - u) / s

where u is the mean of the training samples or zero if with_mean=False, and s is the standard deviation of the training samples or one if with_std=False.

```python
# standardizing the dataset
# X_train ko standardise karenge 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#X-train me fit transform ka use karenge 
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# agar X_train ko inverse karna hai to 
# scaler.inverse_transform(X_train)
```

### Now applying Linear Regression algorithm 
### yahan humare pass multiple input features hai to basically ye hoga multiple Linear regression.
### https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

1.) Linear regression dekhne per hume sirf 2 hi parameters milte hai.
to hume ise Hyper parameter Tune kaise kar sakte hai Ridge, Laso, Elastic net ka use kar sakte hai.

2.) Or second trick Cross validation ka use kar sakte hai
cross validation basically internaly training data ko bahoot se divide karta hai basically multiple train, test me divide karke cross validate karenge or har ek cross validation ka different different accuracy aata hai.

3.) or vo sare validatin ko combine karke uska mean nikalte hai to wo humara final output hota hai.
### https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression() # create LinearRegression Object 
regression.fit(X_train, y_train)
```

```python
## cross validation
from sklearn.model_selection import cross_val_score
# yahan humne cv=5 kiya iska matlab 5 different different data ke sath same model trained hoga 
# to 5 different types ke accuracy ayega 
# Cross validation se humne varify kiya hai ki nagitive mean sqaure kitna ayega
mse = cross_val_score(regression, X_train, y_train, scoring = "neg_mean_squared_error", cv =5)
# mse = mean squared error ye pridicted or truth value ka difference hota hai
```

### https://scikit-learn.org/stable/model_selection.html
### https://scikit-learn.org/stable/modules/model_evaluation.html

```python
np.mean(mse) # final Output 
```

```python
## now pridiction karenge test data ke sath 

reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
## Ab prediction correct hai ki nahi isko dekhne ke liye hum ise compair karenge apne truth value ke sath 
## humara truth value y_test hai 
import seaborn as sns
sns.histplot(reg_pred - y_test, kde=True)
```

## Upar plot me variance kum hai -2 to + 2  matlab ki humara model acha hai 


```python
from sklearn.metrics import r2_score 
# adjecent r2_score ke liye koi aisa default function nahi hai 
# ise nikalne ke liye normal formula iska jo hai use apply karna hota hai 
```

```python
score = r2_score(reg_pred,y_test)
```
