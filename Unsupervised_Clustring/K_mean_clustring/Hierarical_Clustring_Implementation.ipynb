```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np 
%matplotlib inline
```

```python
## Import iris dataset
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

#### kyunki yai features ko cluster karna hai.
#### but humko cluster karke 2D graph pe dikhana hai to 4D me Possible nahi hai
#### K-mean clustering and Heirerical Clustering me Feature Scaling Karna Chahiye 
#### to pahele Feature Scaling Karege 
#### phir 4D se hume 2D me Convert karna hoga 




```python
# Standard Scaler ke pahele Train test split kar sakte hai 
# or seprate train data or test data ka StandardScaler nikalna hai 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
#4D se 2D me lane ke liye Dimensionality Reduction ka use karna hai
#Dimensonality Reduction k liye hume use karna hai PCA
#Apply The PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 4D to 2D convert
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
## Aglomerative Clustring
## To construct a dendogram

import scipy.cluster.hierarchy as sc
##plot the dendogram
plt.figure(figsize=(20,7))
plt.title("Dendograms")

#create dendograms
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendogram')
plt.xlabel('sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2,metric='euclidean', linkage ='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1], c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []
# Notice you start at 2 clusters for silhouette coefficient
for k in range(2,11):
    agglo = AgglomerativeClustering(n_clusters = k, metric = 'euclidean', linkage = 'ward')
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
## plotting silhouette score
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhoute Coeffecient")
plt.show()
```
